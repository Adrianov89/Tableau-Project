```python
# Import packages

import pandas as pd
```

```python
# Load dataset

df= pd.read_csv('fifa18_clean.csv')

df.head()
```

```python
# Display countries that have at least 30 players in the dataset. Save them into a new dataframe (df2):

country_counts = df['Nationality'].value_counts()

countries_with_50_players = country_counts[country_counts >= 50].index

df2 = df[df['Nationality'].isin(countries_with_50_players)]

df2
```

```python
# Save only the best 30 players per country in a new dataframe ('top_50_players_per_country'):

top_50_players_per_country = (
    df2.sort_values(by='Overall', ascending=False)
              .groupby('Nationality')
              .head(50)
)

top_50_players_per_country
```

```python
# Export the final dataframe into a csv file to work on it on Tableau:

top_50_players_per_country.to_csv('top_50_players_per_country.csv', index = False)
```
